3. Let f(x) = $x^5 + 5e^{−3x}$

  Compute the zero order, first order and second order Taylor series approximations
  for the function around x0 = 1 and x0 = 2. Plot these functions along with f(x). You need to generate
  two sets of plots - one each for the x0 values.

#####  Answer-3 Taylor Series Approximation 


```python
from numpy import exp, array, linspace

def f(x) : 
  return x**5 + 5*exp(-3*x)

def f_(x) : # First derivative of f at x  
  return 5*x**4 -15*exp(-3*x)

def f__(x) : # Second derivative of f at x 
  return 20*x**3 + 45*exp(-3*x)

def zero_order_approximation(x0,h) : 
  return f(x0 + h - h)

def first_order_approximation(x0,h) : 
  return f(x0) + f_(x0)*h

def second_order_approximation(x0,h) : 
  return f(x0) + f_(x0)*h + (h**2/2)*f__(x0)

h = linspace(-0.5,0.5,num=20)
```

```python
import matplotlib.pyplot as plt

X = linspace(-3,10,num=100)
plt.plot(X,f(X),label='Actual function f(x)',color='g')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 

# Case 1 : x0 = 1 

x0 = 1 

plt.plot(x0+h,zero_order_approximation(x0,h), label='Zero order approx.',ls=':')
plt.plot(x0+h,first_order_approximation(x0,h),label='First order approx.')
plt.plot(x0+h,second_order_approximation(x0,h),label='Second order approx.')
plt.plot(x0+h,f(x0+h),label = 'Actual function f(x) near x0=%d'%x0)
# plt.axvline(x=x0,ymin=0,ls=':')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt 
# Case 2 : x0 = 2 

x0 = 2

plt.plot(x0+h,zero_order_approximation(x0,h), label='Zero order approx.',ls=':')
plt.plot(x0+h,first_order_approximation(x0,h),label='First order approx.')
plt.plot(x0+h,second_order_approximation(x0,h),label='Second order approx.')
plt.plot(x0+h,f(x0+h),label = 'Actual function f(x) near x0=%d'%x0)
plt.legend()
plt.show()
```
