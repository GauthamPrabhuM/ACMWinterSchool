#### 3. Contour of a function


The Rosenbrock function is a commonly used function for testing the performance of optimization algorithms and is computed as follows

$f(x1, x2) = (1 − x_1)^2 + 100(x_2 − x_1^2)^2$

Generate a single plot showing the contours f(x1, x2) = α for α ∈ {0.5, 2, 5,10, 50, 100, 200, 400, 800}.
You may use the range −2 ≤ x1 ≤ 2, −2 ≤ x2 ≤ 2. Also mark the gradient vectors on this plot.

```python
import matplotlib.pyplot as plt 
from numpy import linspace, meshgrid

x1 = linspace(-2,2,num=20)
x2 = linspace(-2,2,num=20)

x1,x2 = meshgrid(x1,x2)

def f(x1,x2):
  return (1-x1)**2 + 100*(x2-x1**2)**2

levels = [0.5,2,5,10,50,100,200,400,800]

# Gradient vector 
[f_x1, f_x2] = [ 2*(x1-1)-400*x1*(x2-x1**2), 200*(x2 - x1**2)]


contour = plt.contourf(x1,x2,f(x1,x2),levels,alpha=0.7)
plt.quiver(x1,x2,f_x1,f_x2)
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour plot of f(x)')
plt.colorbar(contour)
plt.show()
```

```python
contour
```

```python
# Actual surface plot of f(x)
ax = plt.axes(projection='3d')
ax.plot_surface(x1,x2,f(x1,x2))
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Surface plot of f(x1,x2)')
plt.show()
```
