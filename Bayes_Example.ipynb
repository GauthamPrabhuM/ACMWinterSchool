```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = np.linspace(60, 80, 1001)
```

```python
prior = norm.pdf(x, 67, 2)
likelihood = norm.pdf(x, 75, 1)
product = likelihood * prior
dx = x[2]-x[1]
posterior = product/(np.sum(product)*dx)
```

```python
plt.plot(x, prior, 'b-', lw=5, alpha=0.6, label='Prior')
plt.plot(x, likelihood, 'g-', lw=5, alpha=0.6, label='Likelihood' )
plt.plot(x, posterior, 'r-', lw=5, alpha=0.6, label='Posterior' )
plt.legend(['Prior', 'Likelihood', 'Posterior'])
```

```python
print(np.max(posterior))
```
