```python
import random
```

```python
def f(S):
  influenced = S.copy()
  next_influence = S.copy()
  while len(next_influence) > 0 :
    gets_influenced = set()
    for p in next_influence :
      for c in children[p]:
        if c in influenced : continue
        if random.uniform(0,1) < influence[(p,c)]:
          gets_influenced.add(c)
          influenced.add(c)
    next_influence = gets_influenced
  return len(influenced)
```

```python
def f_avg(S):
  total = 0
  for i in range(1000) :
    total += f(S)
  return total/1000
```

```python
def greedy():
  S = set()
  v = set(vertices)
  while len(S) < 3 :
    max_val = 0
    max_p = 0
    for p in v: 
      t = f_avg(S.union({p}))
      if (t > max_val) :
        max_val = t
        max_p = p
    print(max_p)
    S.add(max_p)
    v.discard(max_p)
  print(S,max_val)
```
